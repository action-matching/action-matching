```python
import torch
import numpy as np

import sys
sys.path.append('../')
from core.hmodel import *
from core.losses import *
from models.mlp import MLP
from models import ema
from utils.train_utils import *
from utils.plot_utils import *
from utils.eval_utils import *

from ml_collections import config_dict
from tqdm.auto import tqdm, trange
```

```python
config_path_am = '../checkpoint/8724567/am_hydrogen.config'
config_path_sm = '../checkpoint/8724481/sm_hydrogen.config'
config_path_ssm = '../checkpoint/8724464/ssm_hydrogen.config'
config = torch.load(config_path_ssm)
```

```python
device = torch.device('cuda')
config.data.batch_size = int(1e4)
model, data_gen, loss, _ = prepare_hydrogen(device, config)
use_ema = True

state = torch.load(config.model.checkpoints[2])
model.load_state_dict(state['model'], strict=True)
if use_ema:
    ema_ = ema.ExponentialMovingAverage(model.parameters(), decay=config.eval.ema)
    ema_.load_state_dict(state['ema'])
    ema_.copy_to(model.parameters())

model.eval()
```

```python
evaluate(model, ema_, data_gen, device, config)
```

```python
def plot_frame(x, x_gt, kde=False):
    fig, ax = plt.subplots(2,3, figsize=(16,10))
    if kde:
        plot_samples_kde(x_gt, axes=ax[0])
        plot_samples_kde(x, axes=ax[1])
    else:
        plot_samples(x_gt, bins=40, axes=ax[0])
        plot_samples(x, bins=40, axes=ax[1])
    for j in range(3):
        ax[0,j].set_title('training data', fontsize=15)
        ax[1,j].set_title(f'samples from {config.model.method}', fontsize=15)
    plt.draw()
    fig.tight_layout()
    return fig
```

```python
x = data_gen.samples[0].to(device)
t = torch.zeros([len(x),1]).to(device)
dt = 1.0/config.data.n_steps
for i in trange(config.data.n_steps):
    x = model.propagate(t, x, dt)
    t += dt

plot_frame(x, data_gen.samples[-1]).show()
```
